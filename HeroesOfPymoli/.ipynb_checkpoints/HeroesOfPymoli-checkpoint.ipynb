* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df
```

## Player Count

* Display the total number of players


```python
# calculate number of unique players
player_count = len(purchase_data_df['SN'].unique())

# place into Total Players Summary DataFrame
total_players_df = pd.DataFrame({"Total Players": [player_count]
                                })
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# calculate number of unique items, the average price of those items, the total number of purchases, and the total revenue
unique_items = len(purchase_data_df['Item Name'].unique())
average_price = purchase_data_df['Price'].mean()
number_purchases = len(purchase_data_df['Purchase ID'])
revenue = purchase_data_df['Price'].sum()

# place calculated information in Purchasing DataFrame
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                       "Average Price": [average_price],
                                       "Number of Purchase": [number_purchases],
                                       "Total Revenue": [revenue]
                                })

# convert columns formatting to two decimal places and add dollar sign
purchasing_analysis_df['Average Price'] = purchasing_analysis_df['Average Price'].astype(float).map("${:,.2f}".format)
purchasing_analysis_df['Total Revenue'] = purchasing_analysis_df['Total Revenue'].astype(float).map("${:,.2f}".format)

purchasing_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# drop duplicates for accurate gender counts
gender_demographics_df = purchase_data_df.drop_duplicates(subset='SN', keep='first')

# calculate gender totals and percentages
unique_gen_count = gender_demographics_df['Gender'].value_counts()
gen_percent = (unique_gen_count / unique_gen_count.sum()) * 100

# create data frame with calculated totals
gen_dem_summary_df = pd.DataFrame({"Total Count": unique_gen_count,
                                   "Percentage of Players": gen_percent.astype(float).map("{:,.2f}%".format)
                                })

gen_dem_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# set up GroupBy using Gender as value
gen_purch_analysis_gb = purchase_data_df.groupby(['Gender'])

# calculations for purchase count, avg. purchase price, total purchase value, avg. purchase total per person 
purch_count = gen_purch_analysis_gb['Purchase ID'].count()
total_purch_value = gen_purch_analysis_gb['Price'].sum()
avg_purch_price = total_purch_value / purch_count
avg_purch_price_person = total_purch_value / unique_gen_count

# create DataFrame with calculations
purch_analysis_gen_df = pd.DataFrame({'Purchase Count': purch_count,
                                      'Average Purchase Price': avg_purch_price.astype(float).map("${:,.2f}".format),
                                      'Total Purchase Value': total_purch_value.astype(float).map("${:,.2f}".format),
                                      'Avg Total Purchase Per Person': avg_purch_price_person.astype(float).map("${:,.2f}".format),
                                    })
purch_analysis_gen_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins for the age groups
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999999]
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# 
purchase_data_df['Age Range'] = pd.cut(purchase_data_df["Age"], bins, labels=age_ranges, include_lowest=True)

# drop duplicates for accurate gender counts
age_bin_df = purchase_data_df.drop_duplicates(subset='SN', keep='first')

age_range_gb = age_bin_df.groupby("Age Range")

print(age_range_gb.head())

# age_range_count = age_range_gb.count()
# age_count_perc = age_range_count / len(age_range_gb)

# age_demographics_df = pd.DataFrame({'Total Count': age_range_count,
#                                     'Percentage of Players': age_count_perc,
#                                    })
# age_demographics_df
```
